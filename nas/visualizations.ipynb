```python
import numpy as np
from torchvision import datasets, transforms

import matplotlib.pyplot as plt

# Load CIFAR-10 dataset
transform = transforms.ToTensor()
train_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)

# Get class names
class_names = train_dataset.classes

# Find the first image of each class
first_images = {}
for img, label in train_dataset:
    if label not in first_images:
        first_images[label] = img
    if len(first_images) == len(class_names):
        break

# Plot the first image of each class
plt.figure(figsize=(12, 6))
for i, (label, img) in enumerate(first_images.items()):
    plt.subplot(2, 5, i + 1)
    plt.imshow(np.transpose(img.numpy(), (1, 2, 0)))
    plt.title(class_names[label])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import torch

# Load synthetic CIFAR-10 dataset
images_best = torch.load('./data/synthetic-cifar10/images_best.pt')
labels_best = torch.load('./data/synthetic-cifar10/labels_best.pt')

# Get unique class labels
unique_labels = torch.unique(labels_best).tolist()

# Define the desired class order
desired_order = ['frog', 'truck', 'deer', 'automobile', 'bird', 'horse', 'ship', 'cat', 'dog', 'airplane']

# Map class names to their corresponding indices
class_name_to_index = {name: idx for idx, name in enumerate(class_names)}

# Find the first image of each class in the synthetic dataset
synthetic_first_images = {}
for img, label in zip(images_best, labels_best):
    if label.item() not in synthetic_first_images:
        synthetic_first_images[label.item()] = img
    if len(synthetic_first_images) == len(class_names):
        break

# Reorder the synthetic_first_images dictionary based on the desired order
ordered_synthetic_first_images = {class_name_to_index[name]: synthetic_first_images[class_name_to_index[name]] for name in desired_order}

# Plot the first image of each class in the desired order
plt.figure(figsize=(12, 6))
for i, (label, img) in enumerate(ordered_synthetic_first_images.items()):
    plt.subplot(2, 5, i + 1)
    plt.imshow(np.transpose(img.numpy(), (1, 2, 0)))
    plt.title(class_names[label])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
